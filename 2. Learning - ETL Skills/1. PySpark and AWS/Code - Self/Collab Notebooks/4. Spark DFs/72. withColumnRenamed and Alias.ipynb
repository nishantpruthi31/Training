```

from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Spark Dataframes").getOrCreate()

```

```
df=spark.read.options(header='True',inferSchema='True').csv('StudentData.csv')
df.show()
```

```
df=df.withColumnRenamed("gender","Sex").withColumnRenamed("roll","roll number")  # the best thing is , if it does not find these column , it will skip it out and won't throw exception
df.show()
```

```
from pyspark.sql.functions import col
df.select(col("name").alias("Full Name")).show()  # creting an alias at run time , but actual will be alright
```
