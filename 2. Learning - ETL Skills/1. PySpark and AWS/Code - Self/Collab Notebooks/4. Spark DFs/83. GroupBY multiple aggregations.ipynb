```
# grouping by from multiple columns
```

```
df=spark.read.options(header='True',inferSchema='True').csv('StudentData.csv')
df.show()
```

```


df.groupBy("course","gender").count().show()
```

```
# now if we want to perform multiple aggregations in 1 statement liks count , avg we will import
from pyspark.sql.functions import sum,avg,max,min,mean,count

df.groupBy("course","gender").agg(count('*').alias("count"),sum("marks"),min("marks"),max("marks")).show()   # we can use alias to give each column a name , else we will see weird names
```
