```

from pyspark.sql import SparkSession
from pyspark.sql.functions import col,lit
spark=SparkSession.builder.appName("Spark Dataframes").getOrCreate()

```

```
df=spark.read.options(header='True',inferSchema='True').csv('StudentData.csv')
df.show()
```

```
df=df.withColumn("total_marks",lit(120))
```

```
df=df.withColumn("average",((col("marks")/col("total_marks"))*100))
```

```
dfOOP=df.filter( (df.course=='OOP') & (col("average")>80))
```

```
dfCloud=df.filter( (df.course=='Cloud') & (col("average")>60))
```

```

dfOOP.select("name","marks").show()
```
