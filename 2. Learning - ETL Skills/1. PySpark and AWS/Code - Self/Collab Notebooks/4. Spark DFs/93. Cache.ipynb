```
# if we cache the data using cache and persist transformation , we will save time and all
```

```
df=spark.read.options(header='True',inferSchema='True').csv('StudentData.csv')
df.show()
```

```
df=df.groupBy("course","gender","age").count()
```

```
df=df.withColumn("dummy",col("age")*100)
```

```
df.show()
```

```
df.cache()  # here we saved the data in cache so that it won't start from begining
```

```
df.show()  # it will read from cache , not all over from top
```
