```

from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Read File")
sc = SparkContext.getOrCreate(conf=conf)   # we provide configuration
```

```

rdd = sc.textFile('Sample.txt')
rdd.collect()
```

```

rdd2=rdd.distinct()
rdd2.collect()
```

```
rdd2=rdd.flatMap(lambda x:x.split(" "))  # first we flatmap
rdd3=rdd2.distinct()
rdd3.collect()
```

```
# one line fxn
rdd.flatMap(lambda x:x.split(" ")).distinct().collect()
```
