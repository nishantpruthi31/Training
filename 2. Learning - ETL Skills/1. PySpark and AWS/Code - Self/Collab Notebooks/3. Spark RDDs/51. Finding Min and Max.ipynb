```

from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Read File")
sc = SparkContext.getOrCreate(conf=conf)   # we provide configuration
```

```

rdd = sc.textFile('movie_ratings.csv')
rdd.collect()
```

```
x='Hi there,2'
x.split(',')
# for creating a tuple we can do
(x.split(',')[0],x.split(',')[1])
```

```
rdd2=rdd.map(lambda x:(x.split(',')[0],  (int(x.split(',')[1]))))
rdd2.collect()
```

```
rdd3=rdd2.reduceByKey(lambda x,y: x if x<y else y)   # so that minimum will be x
rdd3.collect()   # min rating by a movie
```

```
rdd4=rdd2.reduceByKey(lambda x,y: x if x>y else y)   # so that minimum will be x
rdd4.collect()   # max rating by a movie
```
