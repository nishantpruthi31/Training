```
# Repartition is used to change the number of partitions in RDD
# It will create a new RDD
# rdd.repartition(number_of_partitions)

# Coalesce is used to decrease the number of partitions in RDD
# It will create a new RDD
# rdd.coalesce(number_of_partitions)
# coalesce is only used to decrease the number of partition



```

```

from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Read File")
sc = SparkContext.getOrCreate(conf=conf)   # we provide configuration
```

```

rdd = sc.textFile('sample_words.txt')
rdd.collect()
```

```
rdd.getNumPartitions()  # no of partitions for now
```

```
rdd=rdd.repartition(5)   # we change partition
rdd.getNumPartitions()
```

```
rdd = sc.textFile('sample_words.txt')
rdd=rdd.repartition(5)   # we will process 5 partitions
rdd2=rdd.flatMap(lambda x :x.split(' '))
rdd3=rdd2.map(lambda x:(x,1))

```

```
rdd3.saveAsTextFile('output/5_Part_Output')  # output will be created 5 partitions
```

```
# now let's use coleasce also

rdd = sc.textFile('sample_words.txt')
rdd=rdd.repartition(5)   # we will process 5 partitions
rdd2=rdd.flatMap(lambda x :x.split(' '))
rdd3=rdd2.map(lambda x:(x,1))
rdd3=rdd3.coalesce(3)   # we are decreasing partitions here for output file
rdd3.saveAsTextFile('output/3_Part_Output')  # output will be created 3 partitions
```

```
rdd=sc.textFile('output/5_Part_Output')
rdd.getNumPartitions()   # even the output file was having 5 partitions , it combined them all while reading and then by default 2 partition.
```
