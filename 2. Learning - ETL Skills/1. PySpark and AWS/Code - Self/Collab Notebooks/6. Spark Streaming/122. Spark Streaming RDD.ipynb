```python
# SPARK Streaming import some

from pyspark.streaming import StreamingContext
from pyspark import SparkConf,SparkContext
conf=SparkConf().setAppName("Streaming")
sc=SparkContext.getOrCreate(conf=conf)

ssc=StreamingContext(sc,1)   # 2nd argument is duration after which it starts reading from directory
```

```python
rdd=ssc.textFileStream("Files/")   # we get and RDD
```

```python
rdd=rdd.map(lambda x:(x,1))
rdd=rdd.reduceByKey(lambda x,y:x+y)
rdd.pprint()   # here rdd.collect() won't work as it is data stream
ssc.start()
ssc.awaitTerminationOrTimeout(100)   # i.e. if after 100 s we don't get any file , then stop job
```

```python
ssc.stop()  # we will use this to stop dac , in case it throws exception
```
